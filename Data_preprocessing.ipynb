```python
import os
from PIL import Image
from tqdm import tqdm

folders = ['train','test']

new_size = (512,512) 

for folder in folders:
    files = os.listdir(f'../open/{folder}')
    for file in tqdm(files):
        image = Image.open(f'../open/{folder}/{file}')
        resized_image = image.resize(new_size)
        if folder == 'train':
            resized_image.save(f'../open/new_train/{file}')
        else:
            resized_image.save(f'../open/new_test/{file}')
```
