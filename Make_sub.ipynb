```python
import pandas as pd

# 텍스트 파일 경로
txt_file_path = '../Sub/test_score2.txt'

# 텍스트 파일을 데이터프레임으로 읽어오기
# 이 때, 데이터가 공백 또는 탭으로 구분되어 있다고 가정합니다.
# 만약 데이터 구분자가 다르다면 delimiter 매개변수를 해당 구분자로 설정하세요.
df = pd.read_csv(txt_file_path, delimiter='\t',header=None)  # 탭으로 구분된 경우

# 데이터프레임 확인
df.columns = ['img_name','mos']
df['img_name'] = df['img_name'].apply(lambda x : x.split('.')[0])
df
```

```python
submission = pd.merge(sub_df,df,on='img_name')
submission
```

```python
submission = pd.merge(sub_df,df,on='img_name')
df_sub = submission[['img_name','mos_y','comments']]
df_sub.columns = ['img_name','mos','comments']
df_sub
```
